## Laps

```python
session_keys = session_df['session_key'].unique().tolist()
```

```python
laps_df = pd.DataFrame()

for session_key in session_keys:
    api_url = f'https://api.openf1.org/v1/laps?session_key={session_key}'

    response = urlopen(api_url)
    data = json.loads(response.read().decode('utf-8'))
    
    session_df = pd.DataFrame(data)

    laps_df = pd.concat([laps_df, session_df], ignore_index=True)
```

```python
laps_df
```
